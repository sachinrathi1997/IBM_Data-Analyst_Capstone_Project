### Based on the boxplot of the column "Age", where do you see the outliers?

```python
# Open a connection to the database
conn = sqlite3.connect("m4_survey_data.sqlite")
cursor = conn.cursor()

# Query the data from the Age column in the master table
cursor.execute("SELECT Age FROM master")
data = cursor.fetchall()

# Close the connection
conn.close()

# Convert the data to a list of ages
ages = [row[0] for row in data if row[0] is not None]

# Convert the list to a numpy array for easier calculations
ages_array = np.array(ages)

# Calculate Q1, Q3, and IQR
Q1 = np.percentile(ages_array, 25)
Q3 = np.percentile(ages_array, 75)
IQR = Q3 - Q1

# Define the outlier thresholds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers_above = ages_array[ages_array > upper_bound]
outliers_below = ages_array[ages_array < lower_bound]

# Print outliers and thresholds
print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
print("Lower bound for outliers:", lower_bound)
print("Upper bound for outliers:", upper_bound)
print("Outliers above Q3:", outliers_above)
print("Outliers below Q1:", outliers_below)

# Determine if there are any outliers
if len(outliers_above) > 0 or len(outliers_below) > 0:
    print("There are outliers in the data.")
else:
    print("There are no outliers in the data.")
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
import sqlite3
import matplotlib.pyplot as plt

# Open a connection to the database
conn = sqlite3.connect("m4_survey_data.sqlite")
cursor = conn.cursor()

# Query the data from Age and WorkWeekHrs columns in the master table
cursor.execute("SELECT Age, WorkWeekHrs FROM master")
data = cursor.fetchall()

# Close the connection
conn.close()

# Filter out rows with None values in either column
ages = []
work_week_hrs = []

for row in data:
    age, work_week_hr = row
    if age is not None and work_week_hr is not None:
        ages.append(age)
        work_week_hrs.append(work_week_hr)

# Plot the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(ages, work_week_hrs, alpha=0.5)
plt.title('Scatter Plot of Age vs. WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('Work Week Hours')
plt.grid(True)
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
import sqlite3
import matplotlib.pyplot as plt

# Open a connection to the database
conn = sqlite3.connect("m4_survey_data.sqlite")
cursor = conn.cursor()

# Query the data from WorkWeekHrs, CodeRevHrs, and Age columns in the master table
cursor.execute("SELECT WorkWeekHrs, CodeRevHrs, Age FROM master")
data = cursor.fetchall()

# Close the connection
conn.close()

# Filter out rows with None values in any of the columns
work_week_hrs = []
code_rev_hrs = []
ages = []

for row in data:
    work_week_hr, code_rev_hr, age = row
    if work_week_hr is not None and code_rev_hr is not None and age is not None:
        work_week_hrs.append(work_week_hr)
        code_rev_hrs.append(code_rev_hr)
        ages.append(age)

# Plot the bubble plot
plt.figure(figsize=(12, 8))
plt.scatter(work_week_hrs, code_rev_hrs, s=[age * 10 for age in ages], alpha=0.5, c=ages, cmap='viridis', edgecolors='w')
plt.title('Bubble Plot of WorkWeekHrs vs. CodeRevHrs (Bubble Size: Age)')
plt.xlabel('Work Week Hours')
plt.ylabel('Code Review Hours')
plt.colorbar(label='Age')  # Show color scale
plt.grid(True)
plt.show()
```

### Based on the scatter plot of Age and WorkWeekHrs , what kind of  correlation do you see between them?     


```python
import sqlite3
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr

# Open a connection to the database
conn = sqlite3.connect("m4_survey_data.sqlite")
cursor = conn.cursor()

# Query the data from the Age and WorkWeekHrs columns in the master table
cursor.execute("SELECT Age, WorkWeekHrs FROM master")
data = cursor.fetchall()

# Close the connection
conn.close()

# Extract Age and WorkWeekHrs values, excluding rows with None values
ages = []
work_week_hrs = []
for row in data:
    age, work_hrs = row
    if age is not None and work_hrs is not None:
        ages.append(age)
        work_week_hrs.append(work_hrs)

# Convert lists to numpy arrays for easier calculations
ages_array = np.array(ages)
work_week_hrs_array = np.array(work_week_hrs)

# Ensure arrays are of the same length
if len(ages_array) != len(work_week_hrs_array):
    raise ValueError("Age and WorkWeekHrs arrays are not the same length.")

# Calculate the Pearson correlation coefficient
correlation_coefficient, _ = pearsonr(ages_array, work_week_hrs_array)

# Print the correlation coefficient
print("Pearson correlation coefficient:", correlation_coefficient)

# Plot the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x=ages_array, y=work_week_hrs_array, alpha=0.5)
plt.title('Scatter Plot of Age vs. WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.grid(True)
plt.show()

# Interpret the correlation coefficient
if correlation_coefficient > 0.7:
    correlation_description = "Very High Positive Correlation"
elif correlation_coefficient < -0.7:
    correlation_description = "Very High Negative Correlation"
elif -0.3 < correlation_coefficient < 0.3:
    correlation_description = "No Correlation"
else:
    correlation_description = "Difficult to Clearly Identify"

print("Correlation Description:", correlation_description)
```

### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
import sqlite3
import matplotlib.pyplot as plt
from collections import Counter

# Open a connection to the database
conn = sqlite3.connect("m4_survey_data.sqlite")
cursor = conn.cursor()

# Query the data from the DatabaseDesireNextYear table
cursor.execute("SELECT DatabaseDesireNextYear FROM DatabaseDesireNextYear")
data = cursor.fetchall()

# Close the connection
conn.close()

# Extract database names and count occurrences
databases = [row[0] for row in data if row[0] is not None]
database_counts = Counter(databases)

# Get the top 5 databases
top_databases = database_counts.most_common(5)
labels, sizes = zip(*top_databases)

# Plot the pie chart
plt.figure(figsize=(10, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

### Stacked Charts


```python
import sqlite3
import matplotlib.pyplot as plt
import numpy as np

# Open a connection to the database
conn = sqlite3.connect("m4_survey_data.sqlite")
cursor = conn.cursor()

# Query the data for the age group 30 to 35
cursor.execute("""
    SELECT WorkWeekHrs, CodeRevHrs
    FROM master
    WHERE Age BETWEEN 30 AND 35
""")
data = cursor.fetchall()

# Close the connection
conn.close()

# Extract values and handle None values
work_week_hrs = [row[0] for row in data if row[0] is not None]
code_rev_hrs = [row[1] for row in data if row[1] is not None]

# Calculate median values
median_work_week_hrs = np.median(work_week_hrs) if work_week_hrs else 0
median_code_rev_hrs = np.median(code_rev_hrs) if code_rev_hrs else 0

# Data for plotting
categories = ['WorkWeekHrs', 'CodeRevHrs']
values = [median_work_week_hrs, median_code_rev_hrs]

# Plot the stacked bar chart
plt.figure(figsize=(10, 6))
plt.bar(categories, values, color=['skyblue', 'orange'])
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30 to 35')
plt.xlabel('Metric')
plt.ylabel('Median Hours')
plt.show()
```

```python
import sqlite3
import pandas as pd

# Open a connection to the database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Define the SQL query
QUERY = """
SELECT * 
FROM DatabaseWorkedWith
ORDER BY Respondent
"""

# Execute the query and read the results into a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Close the database connection
conn.close()

# Define the list to store unique respondents
uniques = []

# Group by Respondent and count the number of databases each respondent has worked with
counts = df.groupby('Respondent').count()

# Identify respondents who have worked with only one database
for i in range(len(counts)):
    if counts['DatabaseWorkedWith'].values[i] == 1:
        uniques.append(counts.index[i])

# Count the number of respondents who have worked with MySQL only
uniq_mysql = 0
for j in range(len(df)):
    if df['Respondent'][j] in uniques:
        if df['DatabaseWorkedWith'][j] == 'MySQL':
            uniq_mysql += 1

print(uniq_mysql)
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
import sqlite3
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Open a connection to the database
conn = sqlite3.connect("m4_survey_data.sqlite")
cursor = conn.cursor()

# Query the data for the age group 45 to 60
cursor.execute("""
    SELECT Age, ConvertedComp
    FROM master
    WHERE Age BETWEEN 45 AND 60
""")
data = cursor.fetchall()

# Close the connection
conn.close()

# Convert the data to a DataFrame for easier processing
df = pd.DataFrame(data, columns=['Age', 'ConvertedComp'])

# Group by Age and calculate the median ConvertedComp for each age
median_comp_by_age = df.groupby('Age')['ConvertedComp'].median()

# Plot the line chart
plt.figure(figsize=(12, 6))
plt.plot(median_comp_by_age.index, median_comp_by_age.values, marker='o', linestyle='-', color='b')
plt.title('Median ConvertedComp for Ages 45 to 60')
plt.xlabel('Age')
plt.ylabel('Median ConvertedComp')
plt.grid(True)
plt.xticks(np.arange(45, 61, 1))  # Set x-ticks to show every year in the range
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here

import sqlite3
import matplotlib.pyplot as plt
from collections import Counter

# Open a connection to the database
conn = sqlite3.connect("m4_survey_data.sqlite")
cursor = conn.cursor()

# Query the data from the MainBranch column in the master table
cursor.execute("SELECT MainBranch FROM master")
data = cursor.fetchall()

# Close the connection
conn.close()

# Extract MainBranch values and count occurrences
main_branch_values = [row[0] for row in data if row[0] is not None]
main_branch_counts = Counter(main_branch_values)

# Data for plotting
branches, counts = zip(*main_branch_counts.items())

# Plot the horizontal bar chart
plt.figure(figsize=(10, 7))
plt.barh(branches, counts, color='skyblue')
plt.title('Distribution of Main Branches')
plt.xlabel('Number of Respondents')
plt.ylabel('Main Branch')
plt.grid(True, axis='x')
plt.show()
```
